```python
# Performing Feature selection
films = data[['id', 'original_title','cast','overview', 'genres',]]
print(films.columns)
```

```python
films['tags']= films['overview'] + films['genres']
films
```

```python
movies_data = films.drop(columns= ['genres','overview'])
movies_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Performing Feature selection
cv = CountVectorizer(max_features=10866, stop_words='english')
cv
```

```python
cv_matrix = cv.fit_transform(movies_data['tags'].values.astype("U")).toarray()
cv_matrix
```

```python
cv_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
value = cosine_similarity(cv_matrix)
```

```python
value
```

```python
def recommend(movies):
    # Ensure movie title exists in the DataFrame
    if movies not in movies_data['original_title'].values:
        print(f"Movie '{movies}' not found in the dataset.")
        return

    # Get the index of the movie title
    index = movies_data[movies_data['original_title'] == movies].index[0]

    # Ensure 'value' (assuming it's the CV matrix) is available and contains data
    if not isinstance(value, list) or len(value) == 0:
        print("CV matrix 'value' is empty or not properly defined.")
        return

    distance = sorted(list(enumerate(value[index])), reverse=True, key=lambda x: x[1])
    for i in distance[1:11]:  # Start from 1 to skip the input movie itself
        print(movies_data.iloc[i[0]].original_title)


```

```python
recommend('X-men')
```
